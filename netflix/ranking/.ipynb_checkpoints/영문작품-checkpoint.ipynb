```python
import pandas as pd
data =pd.read_csv('final_data.csv', encoding='utf-8')
data
```

```python
def clean_English(data):
    
    clean_str = data.str.replace("[^a-zA-Z]", " ")
    clean_str = clean_str.apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
    clean_str = clean_str.apply(lambda x: x.lower())
    return clean_str
```

```python
data = data.astype(str)
data
```

```python
data['imdb_review'] = clean_English(data['r1']) + clean_English(data['r2']) + clean_English(data['r3']) + clean_English(data['r4']) + clean_English(data['r5'])
data
```

```python
def get_jaccard_sim(str1, str2): 
    l = []
    a = set(str1.split())
    for i in range(len(str2)):
        b = set(str2[i].lower().split())
        c = a.intersection(b)
        result = float(len(c)) / (len(a) + len(b) - len(c))
        if result > (len(a)/len(str2[i])+0.17):
            if result == 1:
                return str2[i]
            l.append(str2[i])
    return l
```
