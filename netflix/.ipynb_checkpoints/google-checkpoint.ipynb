```python
def url_Google(key_word, page_number):
    wd.get('https://www.google.com'+'/search?q='+key_word+'&start='+
          page_number)
    rcv_data = wd.page_source
    soupData = BeautifulSoup(rcv_data, 'html.parser')
    cite = soupData.findAll('div', attrs={'class':'yuRUbf'})
    cite = list(cite)
    cite_url = []
    for i in range(len(cite)):
        left_index_url = str(cite[i]).find('href="') + 6
        right_index_url = str(cite[i])[left_index_url:].find('\"')
        cite_url.append(str(cite[i])[left_index_url:left_index_url+right_index_url])
    return cite_url
        
```

```python
def google_Crwl(cite_url):
    search_categoty=['naver', 'daum', 'tistory', 'fmokorea', 'realfolkblues']
    head = []
    body = []
    for i in range(len(cite_url)):
        for j in range(len(cite_url[i])):
            if cite_url[i][j].find(search_categoty[0]) != -1:
                wd.get(cite_url[i][j])
                rcv_data = wd.page_source
                soupData = BeautifulSoup(rcv_data, 'html.parser')
                
                find_class_name_left_index = str(soupData).find('se-fs-')
                find_class_name_right_index = str(soupData)[find_class_name_left_index:].find('\"')
                class_name = str(soupData)[find_class_name_left_index:find_class_name_left_index +
                                                                                  find_class_name_right_index]
                
                crwl_text = naver_title = soupData.findAll('span', attrs={'class':class_name})
                head.append(crwl_text[0].text)
                content = crwl_text[1].text
                for k in range(2,len(crwl_text)):
                    content += ' ' + crwl_text[k].text
                body.append(content)
                
    crwl = {'title':head, 'content':body}        
    return crwl
```
